```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt

# Load the saved XGBoost models:
clf_model = joblib.load('xgb_classifier.pkl')   # Classifier for downtime status
reg_model = joblib.load('xgb_regressor_time_to_event.pkl')  # Time-to-event regressor (predicts offset in seconds)

# Get the expected feature names (they must match what was used during training)
expected_features_clf = clf_model.feature_names_in_
expected_features_reg = reg_model.feature_names_in_

# Assume all_sheets is a dictionary of sheet_name: DataFrame,
# for example:
# all_sheets = pd.read_excel("TestFile.xlsx", sheet_name=None)

final_predictions = []

for sheet_name, df in all_sheets.items():
    # Convert 'Timestamp' to datetime and drop rows with invalid dates
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
    df = df.dropna(subset=['Timestamp'])
    
    # Create 'timestamp_numeric' if missing (not strictly needed for time-to-event but kept for consistency)
    if 'timestamp_numeric' not in df.columns:
        df['timestamp_numeric'] = df['Timestamp'].apply(lambda x: x.timestamp())
    
    # Prepare classifier input: reindex DataFrame to match training features and fill missing values with 0
    X_test_clf = df.reindex(columns=expected_features_clf, fill_value=0).fillna(0)
    
    # Get predicted probabilities for downtime (class 1)
    downtime_probs = clf_model.predict_proba(X_test_clf)[:, 1]
    
    # Attach the predicted probabilities to the DataFrame
    df['downtime_prob'] = downtime_probs
    
    # For debugging/inspection, print probability stats and show a histogram
    print(f"\n=== {sheet_name} Probability Stats ===")
    print(f"Min: {df['downtime_prob'].min()}, Max: {df['downtime_prob'].max()}, Mean: {df['downtime_prob'].mean()}")
    plt.hist(df['downtime_prob'], bins=20)
    plt.title(f'Downtime Probability Distribution - {sheet_name}')
    plt.xlabel('Probability')
    plt.ylabel('Count')
    plt.show()
    
    # Sort rows by downtime probability in descending order
    df_sorted = df.sort_values(by='downtime_prob', ascending=False)
    
    # Pick the top candidate row (the one with the highest downtime probability)
    if not df_sorted.empty:
        top_candidate = df_sorted.iloc[0:1].copy()  # a one-row DataFrame
    else:
        top_candidate = pd.DataFrame()
    
    # If no candidate is found, prepare a placeholder
    if top_candidate.empty:
        sheet_predictions = pd.DataFrame({
            'Sheet': [sheet_name],
            'Index': [None],
            'Downtime_Probability': [None],
            'Predicted_Downtime': ["No downtime predicted"]
        })
    else:
        # Prepare regressor input for the top candidate using its expected features
        X_test_reg = top_candidate.reindex(columns=expected_features_reg, fill_value=0).fillna(0)
        # Predict the time-to-downtime offset (in seconds)
        time_offset = reg_model.predict(X_test_reg)[0]
        # Get the current Timestamp of the candidate row
        current_time = top_candidate['Timestamp'].iloc[0]
        # Compute the predicted downtime by adding the offset (converted to a timedelta) to the current Timestamp
        predicted_downtime = current_time + pd.to_timedelta(time_offset, unit='s')
        
        sheet_predictions = pd.DataFrame({
            'Sheet': [sheet_name],
            'Index': top_candidate.index,
            'Downtime_Probability': top_candidate['downtime_prob'],
            'Predicted_Downtime': [predicted_downtime]
        })
    
    final_predictions.append(sheet_predictions)

# Combine predictions from all sheets
if final_predictions:
    final_df = pd.concat(final_predictions, ignore_index=True)
    # Print the output in the requested format for each scenario
    for sheet in final_df['Sheet'].unique():
        row = final_df[final_df['Sheet'] == sheet].iloc[0]
        print(f"Most probable downtime for {sheet} : {row['Predicted_Downtime']}")
    final_df.to_csv('Ranked_Downtime_TimeToEvent.csv', index=False)
else:
    print("No predictions across any sheets.")
```
